#### **`ÏΩîÎî©ÏòÅÏó≠`**

````python
# ÏΩîÎî© ÏòÅÏó≠
import this

# Î∞©Î≤ï 1
import os

txt_list1 = []
png_list1 = []

a = os.walk('../source/43-1')

for root, dirnames, filenames in os.walk('../source/43-1'):
    # ÌôïÏù∏
    # print(root, '>>>', dirnames, '>>>', filenames)

    for f in filenames:
        ext = f.split('.')[-1]
        
        if ext == 'txt':
            txt_list1.append(f)
        if ext == 'png':
            png_list1.append(f)


print(f'TXT file info : {txt_list1}, Count : {len(txt_list1)}')
print(f'PNG file info : {png_list1}, Count : {len(png_list1)}')

# Î∞©Î≤ï 2
import glob

txt_list2 = glob.glob('../source/43-1/**/*.txt', recursive=True)
png_list2 = glob.glob('../source/43-1/**/*.png', recursive=True)

print(f'TXT file info : {txt_list2}, Count : {len(txt_list2)}')
print(f'PNG file info : {png_list2}, Count : {len(png_list2)}')
````
